```python
a = np.arange(40,50)
b = np.arange(50,60)

plt.scatter(a,b,c='r')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('2D line')
plt.savefig('test.png')
plt.show()
```

```python
plt.plot(a,b,'g*--')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('2D line')
plt.show()
```

```python
# subplot

plt.subplot(2,2,1)
plt.plot(a,b,'b--')
plt.subplot(2,2,2)
plt.plot(a,b,'g*--')
plt.subplot(2,2,3)
plt.scatter(a,b,c='g')
plt.subplot(2,2,4)
plt.scatter(a,b,c='b')

plt.show()
```

```python
np.pi
```

```python
# sin plot

x = np.arange(0,4*np.pi,0.1)
y = np.sin(x)

plt.plot(x,y,'r--')
plt.title('sin plot')
plt.show()
```

```python
# sin,cos,tan

x = np.arange(0,5*np.pi,0.1)
X = np.arange(0,5*np.pi,5)

y_sin = np.sin(x)
y_cos = np.cos(x)
y_tan = np.tan(X)

plt.subplot(3,1,1)
plt.plot(x,y_sin,'g--')
plt.title('sin')

plt.subplot(3,1,2)
plt.plot(x,y_cos,'b--')
plt.title('cos')

plt.subplot(3,1,3)
plt.plot(X,y_tan,'b--')
plt.title('tan')


plt.show()
```

# error bar

```python
import numpy as np 
import matplotlib.pyplot as plt


fig = plt.Figure()
x = np.arange(10)
y = 2.5*np.sin(x/20*np.pi)
yree = np.linspace(0.0,0.2,10)

plt.errorbar(x,y+3,yerr=yree,label='my 1st line')
plt.errorbar(x,y+2,yerr=yree,uplims=True,label='my 2nd line')
plt.errorbar(x,y+1,yerr=yree,lolims=True,label='my 3rd line')

plt.legend(loc='lower right')

plt.errorbar(x,y,yerr=yree,lolims=True,uplims=True,label='4th line')

plt.show()
```

# grid

```python
x = np.linspace(0,1,500)
y = np.sin(4*np.pi*x)*np.exp(-5*x)

fig,ax = plt.subplots()

ax.fill(x,y,zorder=10)
ax.grid(True,zorder=5)
plt.show()
```

```python
x = np.linspace(0, 1, 500)
y = np.sin(4 * np.pi * x) * np.exp(-5* x)

fig, ax = plt.subplots()

ax.fill(x, y, zorder=10)
ax.grid(True, zorder=5)
plt.show()
```

# 3D 

```python
from mpl_toolkits.mplot3d import Axes3D
from mpl_toolkits.mplot3d import axes3d
import matplotlib.pyplot as plt

fig = plt.figure() # error Figure() will not show u the output its ust the instance of the figure

ax = fig.gca(projection='3d')

#data
theta = np.linspace(-4*np.pi,4*np.pi,100)
z = np.sin(theta)
r = z**2+0
x = r*np.sin(theta)

y = r*np.cos(theta)

ax.plot(x,y,z,label='3d cur')
ax.legend()
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_subplot(,projection='3d')# (1,1) as one subplot and 1 indei the posi abb_subplot(ax)--> n,1
#data

x,y,z = axes3d.get_test_data(0.05)

ax.plot_wireframe(x,y,z)
plt.show()
```
